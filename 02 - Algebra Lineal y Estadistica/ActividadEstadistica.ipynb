```python
import numpy as np

import matplotlib.pyplot as plt

# Generate 1,000,000 random data points
data = np.random.normal(loc=0, scale=2.5, size=1000000)

# Plot line graph
plt.figure(figsize=(10, 5))
plt.plot(data)
plt.title('Line Graph')
plt.xlabel('Index')
plt.ylabel('Data')
plt.show()

# Plot bar graph
plt.figure(figsize=(10, 5))
plt.hist(data, bins=100, edgecolor='black')
plt.title('Bar Graph')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.show()

# CDF
plt.figure(figsize=(10, 5))
plt.hist(data, bins=100, cumulative=True, density=True, edgecolor='black')
plt.title('CDF')
plt.xlabel('Data')
plt.ylabel('Cumulative Probability')
plt.show()

# PDF
plt.figure(figsize=(10, 5))
plt.hist(data, bins=100, density=True, edgecolor='black')
plt.title('PDF')
plt.xlabel('Data')
plt.ylabel('Probability')
plt.show()
```

```python
import numpy as np

# Generate 1,000,000 random data points using the Poisson distribution
data = np.random.poisson(lam=5, size=1000000)

# Plot distribution graph
plt.figure(figsize=(10, 5))
plt.hist(data, bins=100, edgecolor='black')
plt.title('Distribution Graph')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.show()

# Plot line graph
plt.figure(figsize=(10, 5))
plt.plot(data)
plt.title('Line Graph')
plt.xlabel('Index')
plt.ylabel('Data')
plt.show()

# CDF
plt.figure(figsize=(10, 5))
plt.hist(data, bins=100, cumulative=True, density=True, edgecolor='black')
plt.title('CDF')
plt.xlabel('Data')
plt.ylabel('Cumulative Probability')
plt.show()

# PDF
plt.figure(figsize=(10, 5))
plt.hist(data, bins=100, density=True, edgecolor='black')
plt.title('PDF')
plt.xlabel('Data')
plt.ylabel('Probability')
plt.show()
```
