#### Contexto
En 1973, el Dr. Alexander Sawchuk del Instituto de tratamiento de señal e imágenes de la [Universidad del Sur de California](https://www.usc.edu) requería una buena imagen para un artículo de conferencia, querían algo sorprendente para asegurar un buen contraste, y algo donde hubiera un rostro humano. 

Un colaborador llevaba un número de noviembre de 1972 de la revista *Playboy*, de la cual decidieron arrancar el tercio superior del póster central para digitalizarlo en una imagen de *512×512* pixeles. La imagen correspondía a la modelo llamada Lenna, cuyo nombre verdadero es Lena Sjööblom (de Suecia), la cual se convirtió en una referencia como imagen de prueba para los algoritmos de procesamiento de imágenes en todo el mundo.

En enero de 1996, la ***IEEE Transactions on Image Processing*** la cual es una de las revistas científicas más importantes a nivel mundial sobre el procesamiento de imágenes, a través de su editor publicó una nota en relación a la modelo indicando que la ingeniería no debe emplear material de ninguna índole que pueda ser interpretado como una afrenta a las mujeres ya que esto ha generado controversia por mucho tiempo, el artículo puede ser consultado [en esta liga](https://ieeexplore.ieee.org/document/8100841). #### Espacio de Color RGB
Una cámara digital funciona con la combinación de luz roja (R), verde (G) y azul (B), donde la combinación de estos tres espectros (a los que se les llama ***bandas***) forman una imagen conocida como imagen de color verdadero. La cámara digital contiene sensores para cada uno de esos colores primarios **RGB**, y cada uno de ellos genera una imagen la cual mide el nivel de dicho color por medio de una escala definida por una cantidad de bits.

<center><img src="./Imagenes/Canales.png" style="width:690px;height:400px;" class="center"></center>

Comúnmente, las imágenes tiene una resolución radiométrica de 8 bits, esto representa:
$$2^8 = \text{256 combinaciones}$$

Debido a ello, cada uno de los pixeles presentes en una imagen de 8 bits representa un valor entero entre 0 y 255, donde el 0 representa el color ***negro*** y el 255 representa el color ***blanco***, los otros 253 valores son una escala en tonos de grises que se usan para identificar la presencia del color **RBG** en cada una de las bandas, es decir: 
> Un color con un valor 0 indica la ausencia de este color y no emite luz.
> 
> Un color con un valor igual a 255 se visualizará brillante o totalmente saturado de ese color.

Este principio fue en gran medida desarrollado por el Ingeniero Mexicano de orígen tapatío [***Guillermo González Camarena***](https://www.gaceta.unam.mx/guillermo-gonzalez-camarena-el-color-en-la-television), cuyas aportaciones fueron sumamente significativas a través de su patente de 1939 titulada "*Sistema Tricromático Secuencial de Campos*".

Debido a ello, una imagen es una matriz de tres dimensiones que contiene valores enteros (entre 0 y 255 para imágenes de 8 bits). Es importante recalcar que los valores ***no pueden estar fuera de esa escala, no pueden ser flotantes y no pueden ser negativos***.

<center><img src="./Imagenes/Matriz_Imagen.png" style="width:477px;height:400px;" class="center"></center>

#### Combinaciones de Colores
La combinación de los valores de las tres bandas genera un valor en color al que se le conoce como ***Color Verdadero***, esto puede comprenderse mejor por medio de una calculadora de colores **RGB** como la disponible [en esta liga](https://www.w3schools.com/colors/colors_rgb.asp).

Para una imagen **RGB** de 8 bits, la cantidad de posibles combinaciones de colores es:
$$ \text{256 valores} ^ \text{3 bandas} = \text{16,777,216 combinaciones} $$

Lo que significa que puede haber más de 16 millones de colores en una imagen **RGB**.

#### Tamaño en Almacenamiento
Para una imagen de *512 x 512* pixeles en formato **RGB** tenemos lo siguiente:
$$ \text{512 renglones} * \text{512 columnas} * \text{3 bandas} = \text{786,432 pixeles} $$

Considerando que cada pixel corresponde a 8 bits (es decir, 1 byte), tenemos que la imagen abarca:
$$ \text{786,432 bytes} * 8 = \text{786,432,000 bits} $$

Al verificar el tamaño de las imágenes podemos ver que son del rango de 200 a 500 KB, esto es debido a que el formato **JPG** realiza una operación de compresión de la imagen para reducir su tamaño, esto con la finalidad de abarcar menor espacio de almacenamiento. Sin embargo, esta compresión disminuye la resolución de la imagen lo cual puede ser problemático en ciertas aplicaciones. 

## Procesamiento de Imágenes en Python
Para ejemplificar las operaciones vectoriales y matriciales se empleará el procesamiento de imágenes de una manera básica, empleando las librerías ya conocidas como **NumPy** y **Matplotlib**. Adicionalmente se empleará la librería **SciKit-Image** para operaciones básicas empleando imágenes.

#### La Librería SciKit-Image
<center><img src="./Imagenes/Scikit-Image_logo.png" style="width:487px;height:120px;" class="center"></center>

**SciKit-Image** es una librería para procesamiento de imágenes en **Python** que incluye algoritmos para segmentación, transformaciones geométricas, manipulación de espacios de color, análisis, filtrado, morfología, entre otras. 

La documentación de **SciKit-Image** se puede encontrar a través [de esta liga](https://scikit-image.org).

Para importar la librería se empleará la función `from skimage import io`, a partir de ello se podrá hacer uso de los distintos métodos del módulo `io` que se emplean para la lectura y escritura de imágenes en distintos formatos, invocándolo simplemente como `io`. El contenido de este módulo está descrito a detalle en [esta liga](https://scikit-image.org/docs/stable/api/skimage.io.html).

Como primer ejemplo se lee el archivo `Lily.jpg` a través del método `.imread()` de **SciKit-Image**. Posteriormente se muestra empleando el método `.imshow()` de **Matplotib**.

La documentación del método `.imread()` de **SciKit-Image** puede ser consultada [en esta liga](https://scikit-image.org/docs/stable/api/skimage.io.html#skimage.io.imread).

Algunos métodos adicionales empleados:

* El método `.axis()` de **Matplotlib** se emplea para mostrar los ejes en la figura.
* El método `.title()` de **Matplotlib** se emplea para mostrar un título en la figura.
* El método `.show()` de **Matplotlib** se emplea para mostrar la figura.
* El método `.shape` de **SciKit-Image** se emplea para mostrar el tamaño de la figura.

***Nota:*** El punto `.` en la lectura del archivo indica a **Python** que se está emplenado el directorio base para acceder a los archivos. Sin embargo, y principalmente en **Windows**, se puede emplear el prefijo `r` para indicar una ruta en particular, este prefijo se conoce como "*raw string*" y como ejemplo se puede escribir algo similar a:
```python
io.imread(r"C:\Users\villalon\Documents\Maestria\Programacion\Codigos\Datos\Lily.jpg")
```

````python
#Importación de Librerías
import matplotlib.pyplot as plt          
import numpy as np                       
from skimage import io                   

#Lectura de la Imagen desde un archivo JPG
Lily = io.imread("./Imagenes/Lily.jpg")           

#Impresión de la Imagen
plt.imshow(Lily)
plt.axis('off')
plt.title("Lily")
plt.show()

#Impresión de los resultados
print("Dimensiones de la imagen:")
print("Lily.jpg:", Lily.shape)        
````

De igual manera se puede realizar la lectura de varias imágenes, e incluso mostrarlas en un arreglo estilo matricial. Para ello se emplea el método `.subplot(a,b,c)` de **Matplotlib**, cuyos argumentos son:

* El argumento `a` indica la cantidad de renglones que tendrá el arreglo de imágenes.
* El argumento `b` indica la cantidad de columnas que tendrá el arreglo de imágenes.
* El argumento `c` indica la posición en el arreglo donde se colocará la imágen a mostrar con el método `.imshow()`.

````python
#Lectura de las Imágenes desde archivos JPG
Lenna   = io.imread("./Imagenes/Lenna.jpg")
Imagen1 = io.imread("./Imagenes/Imagen1.jpg")
Imagen2 = io.imread("./Imagenes/Imagen2.jpg")
Imagen3 = io.imread("./Imagenes/Imagen3.jpg")

#Impresión de las Imágenes
plt.subplot(1,4,1)
plt.imshow(Lenna), plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(Imagen1), plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(Imagen2), plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(Imagen3), plt.axis('off')

plt.show()      
````

Debido a que una imágen en formato **JPG** contiene tres canales, los cuales corresponden a los niveles de intensidad de los colores ***rojo (R)***, ***verde (G)*** y ***azul (B)***, es posible separar cada uno de esos canales empleando operaciones matriciales.

Al aplicar el método `.shape` a la imagen `Lily.jpg` se obtuvo el resultado `(512, 512, 3)`, esto significa que la imagen:

* Contiene 512 pixeles de alto (renglones).
* Contiene 512 pixeles de ancho (columnas).
* Contiene 3 bandas de información (R, G y B).

El valor `3` corresponde a la **Resolución Espectral** de la imagen, que define la cantidad de bandas de información contenidas en la misma. A partir de ello se puede tener una imagen:

* **Pancromática**: 1 banda.
* **Multiespectral**: 2 a 9 bandas. 
* **Hiperespectral**: Más de 9 bandas.

Al considerar que una imagen es una matriz de tres dimensiones `[a,b,c]`, es posible manipular su contenido de la manera descrita a continuación.

***Nota:*** Las imágenes generadas pueden ser guardadas de manera individual empleando el método `.imsave(a,b,c)` de **Matplotlib**, donde sus tres argumentos son:

* El argumento `a` indica el nombre de destino del archivo a ser guardado.
* El argumento `b` indica la variable que contiene los valores a ser guardados.
* El argumento `c` indica el mapa de colores a ser empleado.

La documentación detallada del método `.imsave()` de **Matplotlib** puede ser consultada [en esta liga](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imsave.html).

````python
#Valores de los Canales contenidos en la imagen RGB
Lily_r = Lily[:,:,0]                 
Lily_g = Lily[:,:,1]                
Lily_b = Lily[:,:,2]                

#Impresión de la Banda R
plt.imshow(Lily_r, cmap="gray")        
plt.axis('off')
plt.title("Canal Rojo (R)")
plt.show()
#plt.imsave('./Guardados/Lily_R.jpg', Lily_r, cmap="gray")

#Impresión de la Banda G
plt.imshow(Lily_g, cmap="gray")        
plt.axis('off')
plt.title("Canal Verde (G)")
plt.show()
#plt.imsave('./Guardados/Lily_G.jpg', Lily_g, cmap="gray")

#Impresión de la Banda B
plt.imshow(Lily_b, cmap="gray")        
plt.axis('off')
plt.title("Canal Azul (B)")
plt.show()
#plt.imsave('./Guardados/Lily_B.jpg', Lily_b, cmap="gray")
````

El manejo de los valores de los pixeles también pueden emplearse para extraer zonas específicas dentro de la imagen. Por ejemplo, para mostrar el ojo derecho de `Lily` se puede indicar la ventana de pixeles que lo contienen:

* Renglones: pixeles del 130 al 190.
* Columnas: pixeles del 150 al 210.

Esto es:

````python
#Valores de los Canales contenidos en la imagen RGB
Lily_ojo = Lily[130:190, 150:210, :]

#Impresión de las Imágenes
plt.imshow(Lily_ojo)        
plt.axis('off')
plt.title("Ojo Derecho de Lily")
plt.show()
#plt.imsave('./Guardados/Lily_Ojo.jpg', Lily_ojo)
````
